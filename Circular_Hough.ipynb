```python
import sys
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/SS urang/800px-Mangga_gedong_gincu_071021-0845_tmo.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (17,17), 0)
buled = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1, 20, param1=100, param2=30, minRadius=75, maxRadius=200)

detek_buled = np.uint16(np.around(buled))
for i in detek_buled[0, :]:
  tengah = (i[0], i[1])
  cv2.circle(img, tengah, 1, (0,100,100), 3)
  radius = i[2]
  cv2.circle(img, tengah, radius, (255, 0, 255), 3 )

cv2_imshow(img)
```
